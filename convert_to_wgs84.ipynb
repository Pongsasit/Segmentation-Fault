```python
import glob
import os
from pathlib import Path
import subprocess
```

```python
base_path = Path('./data/sentinel-2-image')
cwd = os.getcwd()
```

convert jp2 files

```python
paths = glob.glob(os.path.join(cwd, 'data', 'sentinel-2-image', '*', '*', 'IMG_DATA', '*.jp2'))
for path in paths:
    parts = path.split('\\')
    new_parts = list(map(lambda x: x.replace('data', 'data_wgs84'), parts))
    dest_path = '\\'.join(new_parts)
    dest_dir = '\\'.join(new_parts[:-1])
    print(dest_dir)
    if not os.path.exists(dest_dir):
        os.makedirs(dest_dir)
    subprocess.run(f'gdalwarp {path} {dest_path} -t_srs "+proj=longlat +ellps=WGS84"')
```
